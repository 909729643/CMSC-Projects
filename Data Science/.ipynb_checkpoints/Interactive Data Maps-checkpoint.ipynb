```python
!pip install folium
import folium
from folium.map import *
import numpy as np
import requests
import pandas

arrest_table = pandas.read_csv("http://www.hcbravo.org/IntroDataSci/misc/BPD_Arrests.csv")
arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)

arrest_table.head()
#arrest_table.race.unique()
```

# Generate the Interactive Map

This map shows the sample of 0.5% random different arrest cases happened in Baltimore. It shows the locations of the arrest cases, and different colors stand for different sex and race.

```python
import random
map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)
arrest_table_random = arrest_table.sample(frac=0.005)

# Add layers
arrests_sex = FeatureGroup(name='Sex')
arrests_race = FeatureGroup(name='Race')

# Markers for arrest location
# Blue Markers: Male
# Red Markers : Female

# Circles for race
# A - Yellow
# B - Black
# H - Orange
# I - Green
# U - Purple
# W - White
# Nan - None
for i,row in arrest_table_random.iterrows():
    sex = str(row['sex'])
    race = str(row['race'])
    description = str(row['chargeDescription'])
    location = [row['lat'],row['long']]
    if sex == 'M':
        arrests_sex.add_child(folium.Marker(location,popup=description,icon=folium.Icon(color='blue')))
    else:
        arrests_sex.add_child(folium.Marker(location,popup=description,icon=folium.Icon(color='red')))
    if race == 'A':
        arrests_race.add_child(folium.Circle(radius=100,location=location,popup=description,color='yellow',fill=True))
    elif race == 'B':
        arrests_race.add_child(folium.Circle(radius=100,location=location,popup=description,color='black',fill=True))
    elif race == 'H':
        arrests_race.add_child(folium.Circle(radius=100,location=location,popup=description,color='orange',fill=True))
    elif race == 'I':
        arrests_race.add_child(folium.Circle(radius=100,location=location,popup=description,color='green',fill=True))
    elif race == 'U':
        arrests_race.add_child(folium.Circle(radius=100,location=location,popup=description,color='purple',fill=True))
    elif race == 'W':
        arrests_race.add_child(folium.Circle(radius=100,location=location,popup=description,color='white',fill=True))


#Add layers: sex, race
map_osm.add_child(arrests_race)
map_osm.add_child(arrests_sex)
map_osm.add_child(folium.map.LayerControl())
map_osm
```

This map shows that most arrest in Baltimore happened around the center of the city. Most criminals are male instead of female. Also, most of them are blacks.
